```python
import requests
import pandas as pd
import json
from config import api_key

url = "https://api.covidactnow.org/v2/county/29095.timeseries.json?apiKey=" + api_key
print(url)
```

```python
response = requests.get(url).json()
response
```

```python
covid_df = pd.DataFrame(response['actualsTimeseries'])
covid_df = pd.DataFrame(covid_df[['date', 'cases', 'newCases', 'deaths']])
covid_df = covid_df.rename(columns = {'date': 'Date', 'cases': 'Cases', 'newCases': 'Daily Cases', 'deaths': 'Deaths'})
covid_df.fillna(0)
```

```python
# checking format
covid_df.to_csv("covid.csv")
```
